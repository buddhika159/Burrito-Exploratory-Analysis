```python
# cleaning the column names

df.columns = [re.sub("([\(\[]).*?([\)\]])", "", x).strip() for x in df.columns]
df.columns = [x.replace(":","_").strip() for x in df.columns]
# this removes unnecesory () and : in columns
```

```python
df.isnull().sum()
```

```python
session = tt.create_session(config = {"user_content_storage": "./content", "port":9000})
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
burrito_vars_norm = df.loc[:,["Circum","Volume","Length","Mass","Cost"]]
burrito_vars_norm.isnull().sum()
```

```python
# since there are alot of missing values in these columns, lets normalize them and show only the most significant variable name and its value in two new columns
```

```python
# normalize the selected columns 1-10
bnorms = scaler.fit_transform(burrito_vars_norm)*10
df[["Circum_norm","Volume_norm","Length_norm","Mass_norm","Cost_norm"]] = bnorms
```

```python
burrito_variables = pd.melt(df.reset_index(), id_vars = ["Location","Burrito"], value_vars = ["Circum_norm","Volume_norm","Length_norm","Mass_norm","Cost_norm"])
burrito_variables
```

```python
burrito_table = session.read_pandas(df, table_name = "burritos")
burrito_table.head()
```

```python
# cube turns the above table more interactive and allows you to create pivot tables
cube = session.create_cube(burrito_table)
```

```python
h = cube.hierarchies
l = cube.levels
m = cube.measures
```

```python
# create new variables in measures to use in pivot tables

m["five"] = 5
m["four"] = 4
m["four_google"] = 4.25
m["lenXwrap"] = m["Length.MEAN"] * m["Wrap.MEAN"] 
```

```python
session.visualize("exploration 1")
```

```python
session.visualize("scatter plot 1")
```

```python
burrito_var_table = session.read_pandas(burrito_variables, table_name = "burritio_variables", keys = ["Location","Burrito","variable"])
```

```python
burrito_table.join(burrito_var_table)
```

```python
m["value"] = tt.agg.mean(burrito_var_table["value"])
m["aggvalue"] = tt.agg.mean(m["value"], scope = tt.scope.origin(l["Location"],l["Burrito"],l["variable"]))
```
